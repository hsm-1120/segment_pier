```python
def seg(tendon_d, rebar_d, ini_f):

    # create output folder
    storage_path = '../data/output'
    if not os.path.exists(storage_path):
        os.mkdir(storage_path)

    # log file for error
    logFile('error.log', '-noEcho')

    # start build
    print('\033[0;33;45m####<Start Building!>####\033[0m\n')

    wipe()
    model('basic', '-ndm', 3, '-ndf', 6)

    # variables for analysis
    # rebar_d  = 0.02
    # tendon_d = 0.01
    # ini_f    = 80
    ini_ss   = ini_f / (tendon_d ** 2 * np.pi / 4)
    ini_sn   = ini_ss / 1.95e8

#----------------------------------------build node------------------------------------------------#
    print('\033[1;31;47m####<Node Building>####\033[0m\n')
    
    nl = pd.read_excel('../data/input/model_infor/1_node.xlsx')
    for tag, coord_x, coord_y, coord_z in zip(nl.iloc[:, 0], nl.iloc[:, 1], nl.iloc[:, 2], nl.iloc[:, 3]):
        node(tag, coord_x, coord_y, coord_z)
#--------------------------------------boundary-condition---------------------------------------#
    print('\033[1;31;47m####<Boundary Condition>####\033[0m\n')

    fl = pd.read_excel('../data/input/model_infor/2_fixnode.xlsx')
    for tag in nl.iloc[:, 0]:
        fix(tag, 1, 1, 1, 1, 1, 1)

    bl = pd.read_excel('../data/input/model_infor/3_rigidLink.xlsx')
    for inode, jnode in zip(bl.iloc[:, 0], bl.iloc[:, 1]):
        rigidLink('beam', inode, jnode)
    
    geomTransf('Linear', 1, 1.0, 0.0, 0.0)
#----------------------------------------material------------------------------------------------#
    print('\033[1;31;47m####<Material/Fiber Section>####\033[0m\n')
    
    with open('../data/input/model_infor/mat_sec.txt','r') as f:
        exec(f.read())

#--------------------------------------element------------------------------------------------#
    print('\033[1;31;47m####<Element>####\033[0m\n')
    # segment element
    eel = pd.read_excel('../data/input/model_infor/4_elastic.xlsx')

    for tag, inode, jnode, A, E, G, J, Iy, Iz, tran in zip(eel.iloc[:, 0], eel.iloc[:, 1], eel.iloc[:, 2],
                                                           eel.iloc[:, 3], eel.iloc[:, 4], eel.iloc[:, 5],
                                                           eel.iloc[:, 6], eel.iloc[:, 7], eel.iloc[:, 8],
                                                           eel.iloc[:, 9]):
        element('elasticBeamColumn', tag, inode, jnode, A, E, G, J, Iy, Iz, tran)

    trl = pd.read_excel('../data/input/model_infor/5_truss.xlsx')
    for tag, inode, jnode, mat in zip(trl.iloc[:, 0], trl.iloc[:, 1], trl.iloc[:, 2], trl.iloc[:, 3]):
        if mat == 5:
            element('Truss', tag, inode, jnode, tendon_d ** 2 * np.pi / 4, mat)
        else:
            element('Truss', tag, inode, jnode, rebar_d ** 2 * np.pi / 4, mat)

    spl = pd.read_excel('../data/input/model_infor/6_spring.xlsx')
    for tag, inode, jnode, mat1, mat2, mat3 in zip(spl.iloc[:, 0], spl.iloc[:, 1], spl.iloc[:, 2], 
                                                    spl.iloc[:, 3], spl.iloc[:, 4], spl.iloc[:, 5]):
        element('zeroLength', tag, inode, jnode, '-mat', mat1, mat2, mat3, '-dir', 1, 2, 3, '-orient', 0, 0, 1, 1, 0, 0)

    print('\033[0;33;42m####<Finish Building!>####\033[0m')
```

```python
def quasi_static(axial_load, lateral_disp, step, tag):

    folder = '../data/output/quasi_static'
    if not os.path.exists(folder):
        os.mkdir(folder)

    wipeAnalysis()
    timeSeries('Linear', 1)
    pattern('Plain', 1, 1)
    load(tag, 0, 0, axial_load, 0, 0, 0)
    integrator('LoadControl', 0.1)
    system('SparseGeneral')
    test('NormDispIncr', 1.0e-4, 100, 0)
    numberer('RCM')
    constraints('Transformation')
    algorithm('KrylovNewton')
    analysis('Static')
    ok = analyze(10)
    if ok != 0:
        print('\033[1;37;41m####<axial load Failed!>####\033[0m\n')
    else:
        print('\033[1;37;42m####<axial load Succeed!>####\033[0m\n')

    loadConst('-time', 0.0)
    #############################################################
    # rcl = pd.read_excel('data/input/model_infor/7_recorder.xlsx')
    # for tag, item in zip(rcl.iloc[:, 0], rcl.iloc[:, 1]):
    #     savepath = folder + f'/{item}_{tag}.out'
    #     recorder_define(savepath, item, tag)
    # record()
    #############################################################
    ##########################################>-->analysis module
    # cyclic load
    ##########################################>-|->lateral load
    timeSeries('Linear', 2)
    pattern('Plain', 2, 2)
    load(tag, 1, 0, 0, 0, 0, 0)
    system('SparseGeneral', '-piv')
    test('NormDispIncr', 1.0e-4, 100, 0)
    numberer('RCM')
    constraints('Transformation')
    algorithm('KrylovNewton')
    analysis('Static')
    for i in range(step):
        integrator('DisplacementControl', tag, 1, lateral_disp / (10 * step))
        ok1=analyze(10 * (i + 1))
        integrator('DisplacementControl', tag, 1, lateral_disp / (10 * step))
        ok2=analyze(10 * (i + 1))
        integrator('DisplacementControl', tag, 1, lateral_disp / (10 * step))
        ok3=analyze(10 * (i + 1))
        integrator('DisplacementControl', tag, 1, lateral_disp / (10 * step))
        ok4=analyze(10 * (i + 1))
        if (ok1 + ok2 + ok3 + ok4) != 0:
        # if ok1 != 0:
            print('\033[1;37;41m####<No.{} step Failed!>####\033[0m\n'.format(i + 1))
        else:
            print('\033[1;37;42m####<No.{} step Succeed!>####\033[0m\n'.format(i + 1))
    ##########################################>-|->analysis module
```
